# Graph of this DataSet

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_jobs(df, size=(12, 6), style="dark", dashes=False):
    # Set the aesthetic style of the plots
    sns.set(style=style)

    # Create a bar plot for No of Jobs
    plt.figure(figsize=size)
    if dashes:
        experience_plot = sns.barplot(x="Job Name", y="No of Jobs", data=df, palette="viridis", errorbar=None, linewidth=2, edgecolor=".2")
    else:
        experience_plot = sns.barplot(x="Job Name", y="No of Jobs", data=df, palette="viridis")
    
    experience_plot.set_title("Number of Jobs Required for Various Jobs")
    experience_plot.set_xlabel("Job Name")
    experience_plot.set_ylabel("Number of Jobs")
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Show the plot
    plt.show()

# Test the function with default parameters
plot_jobs(df)
```

# Input User Detail

```python
class Resume:
    def __init__(self, name, email, mobile, education, skills, experience, projects, achievements, hobbies):
        # Initialize the Resume object with user information
        self.name = name
        self.email = email
        self.mobile = mobile
        self.education = education
        self.skills = skills
        self.experience = experience
        self.projects = projects
        self.achievements = achievements
        self.hobbies = hobbies

    def generate_markdown(self):
        # Generate Markdown content for the resume
        markdown_text = f"<h1 style=\"text-align:center;\">{self.name}</h1>\n<p style=\"text-align:center;\">Email: {self.email} | Mobile: {self.mobile} </p>\n\n"
        markdown_text += "### Education\n\n---\n\n"
        
        # Add education details to the Markdown content
        for edu in self.education:
            markdown_text += f"- {edu['level']}: {edu['institution']} | {edu['field']} | Score: {edu['score']} | {edu['duration']}.\n\n"

        markdown_text += "### Skills\n\n---\n\n"
        # Add skills to the Markdown content
        markdown_text += f"{', '.join(self.skills)} \n\n"

        markdown_text += "### Experience\n\n---\n\n"
        # Add work experience details to the Markdown content
        for exp in self.experience:
            markdown_text += f"- **{exp['job_role']} ({exp['company_name']})**: {exp['description']}\n"

        markdown_text += "\n### Projects\n\n---\n\n"
        # Add project details to the Markdown content
        for proj in self.projects:
            markdown_text += f"- **{proj['name']}**: {proj['description']}\n"

        markdown_text += "\n### Achievements\n\n---\n\n"
        # Add achievement details to the Markdown content
        for ach in self.achievements:
            markdown_text += f"- {ach}\n"

        markdown_text += "\n### Hobbies\n\n---\n\n"
        # Add hobbies to the Markdown content
        markdown_text += f"{', '.join(self.hobbies)}\n"

        return markdown_text

def get_user_input():
    # Gather user input for creating the resume
    name = input("Enter your name: ")
    email = input("Enter your email: ")
    mobile = input("Enter your mobile number: ")

    print("\nEducation:")
    education = []
    while True:
        # Prompt user to add education details
        edu_input = input("Do you want to add education details? (yes/no): ").lower()
        if edu_input != 'yes':
            break
        level = input("Enter education level (e.g., Graduation(UG/PG), High School): ")
        institution = input(f"Enter the name of the {level} institution: ")
        field = input(f"Enter the field of study at {institution}: ")
        duration = input(f"Enter passing year of {level} at {institution}: ")
        score = input(f"Enter your score (e.g., GPA/Percentage) of {level} at {institution}: ")
        education.append({"level": level, "institution": institution, "field": field, "duration": duration, "score": score})

    skills = input("\nEnter your skills (comma-separated): ").split(',')

    print("\nExperience:")
    experience = []
    while True:
        # Prompt user to add work experience details
        job_role = input("Enter your job role(Experience) (or type 'done' to finish): ")
        if job_role.lower() == 'done':
            break
        exp_company_name = input("Enter the company name: ")
        exp_description = input(f"Enter the description for '{job_role}': ")
        experience.append({"job_role": job_role, "company_name": exp_company_name, "description": exp_description})

    print("\nProjects:")
    projects = []
    while True:
        # Prompt user to add project details
        proj_heading = input("Enter the project Title (or type 'done' to finish): ")
        if proj_heading.lower() == 'done':
            break
        proj_description = input(f"Enter the description for '{proj_heading}': ")
        projects.append({"name": proj_heading, "description": proj_description})

    print("\nAchievements:")
    achievements = []
    while True:
        # Prompt user to add achievement details
        ach_input = input("Enter an achievement detail (or type 'done' to finish): ")
        if ach_input.lower() == 'done':
            break
        achievements.append(ach_input)

    hobbies = input("\nEnter your hobbies (comma-separated): ").split(',')

    return Resume(name, email, mobile, education, skills, experience, projects, achievements, hobbies)


    # Main execution block
user_resume = get_user_input()
markdown_text = user_resume.generate_markdown()
print("Successful! Your resume has been generated.")  # Print confirmation message
```

# Select the JOB

```python
import pandas as pd

# Function to display job names and get user's selection
def select_job(df):
    print("Available Job Titles:")
    for i, job in enumerate(df["Job Name"], 1):
        print(f"{i}. {job}")

    # Prompt user to select a job by entering the corresponding number
    while True:
        try:
            selection = int(input("Select a job by entering the corresponding number: "))
            if 1 <= selection <= len(df):
                selected_job = df.iloc[selection - 1]["Job Name"]
                break
            else:
                print(f"Please enter a number between 1 and {len(df)}.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    return selected_job

# Assume df is the already built-in DataFrame with a column "Job Name"
# Example usage with the existing DataFrame df
selected_job = select_job(df)

# Display the selected job
print(f"\nYou have selected: {selected_job}")
```

# Generate a CV

```python
import requests

def convert_markdown_to_pdf(markdown_content, selected_job, Resume_file="Resume.pdf", engine="weasyprint"):
    # Define CSS styles for the PDF
    cssfile = """
    body{
        padding: 0px;
        margin:0px;
    }
    h1 {
        color: MidnightBlue;
        margin:0px;
        padding:0px;
    }
    h3{
        color: MidnightBlue;
        padding-bottom:0px; 
        margin-bottom:0px; 
    }
    li{
        margin-top:5px;
    }
    """
    
    # API endpoint for converting Markdown to PDF
    url = "https://md-to-pdf.fly.dev"

    # Get profile description based on the selected job
    profile_description = df[df["Job Name"] == selected_job]["Profile Description"].values[0]

    # Insert profile description before the education section
    education_index = markdown_content.find("### Education")
    markdown_content = markdown_content[:education_index] + f"\n### About Me\n\n---\n\n{profile_description}\n\n" + markdown_content[education_index:]

    # Data to be sent in the POST request
    data = {
        'markdown': markdown_content,
        'css': cssfile,
        'engine': engine
    }

    # Send a POST request to the API
    response = requests.post(url, data=data)

    # Check if the response is successful (status code 200)
    if response.status_code == 200:
        # Save the generated PDF to a file
        with open(Resume_file, 'wb') as f:
            f.write(response.content)
        print(f"PDF saved to {Resume_file}")
    else:
        print(f"Error {response.status_code}: {response.text}")


convert_markdown_to_pdf(markdown_text, selected_job)
```

# Analytics

```python
import pandas as pd

# Assuming df is already defined as per your provided data

def compare_skills_and_generate_score(user_resume, selected_job, df):
    # Get the index of the selected job in the DataFrame
    selected_job_index = df[df["Job Name"] == selected_job].index[0]
    
    # Get the skills required for the selected job
    selected_job_skills = set(df["Skills"][selected_job_index].split(', '))

    # Ensure user skills are formatted as a string
    user_skills_str = ', '.join(user_resume.skills)

    # Split the user's skills into individual skills
    user_skills = set(user_skills_str.split(', '))

    # Check if there are any missing skills
    missing_skills = selected_job_skills - user_skills

    if missing_skills:
        print(f"{user_resume.name}, you are missing the following skills for this job:")
        for skill in missing_skills:
            print("-", skill)
        print("Consider learning these skills to improve your suitability for the job.")
    else:
        print(f"Congratulations, {user_resume.name}! You have all the required skills for this job.")

    # Calculate the score by counting matching skills
    score = len(selected_job_skills.intersection(user_skills))

    # Display advice
    print("\nAdvice for this job:")
    print(df["Advice"][selected_job_index])

    # Display tips
    print("\nTips for this job:")
    for i, tip in enumerate(df["Tips"][selected_job_index], start=1):
        print(f"Tip {i}: {tip}")

    return score

# Assuming user_resume and selected_job are defined
# Generate the score
score = compare_skills_and_generate_score(user_resume, selected_job, df)

# Display the score
print(f"\nYour score for the selected job ({selected_job}): {score}")

# Print the name of the resume person
print(f"Resume Name: {user_resume.name}")
```
